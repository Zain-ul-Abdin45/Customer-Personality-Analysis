```python
edu= list(set(data['Education']))
lvl=[]
for i in range(len(edu)):
    lvl.append(data['Education'].to_list().count(edu[i]))
    edu[i]=edu[i]+' ('+str(lvl[i]*100/len(data['Education']))+" %)"
plt.pie(lvl, labels=edu, radius=3)
plt.show()
```

```python
status=list(set(data['Marital_Status']))
stat=[]
for k in range(len(status)):
    stat.append(data['Marital_Status'].to_list().count(status[k]))
plt.bar(x=range(len(status)), height=stat)
plt.xticks(labels=status, ticks=range(len(status)), rotation=90)
plt.show()
```

```python
age=2021-data.Year_Birth.to_numpy()
```

```python
t_age=list(set(age))
age_distribution=[]

for l in range(len(t_age)):
    age_distribution.append(list(age).count(t_age[l]))

plt.figure(figsize=(16,8))
plt.bar(x=range(0,len(t_age)*3, 3), height=age_distribution)
plt.xticks(labels=t_age, ticks=range(0, len(t_age)*3, 3), rotation=90)
plt.show()
```
